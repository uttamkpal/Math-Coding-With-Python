## Harmonograph

```python
n = 1000000
t = np.logspace(np.log10(10), np.log10(500), n)

A = [1,1,1.5,1.5]
d = [.004, .001, .002, .0015]
F = [3, 1, 2, 2.5]

x = A[0]*np.sin(t*F[0])*np.exp(-d[0]*t) + A[1]*np.sin(t*F[1])*np.exp(-d[1]*t)
y = A[2]*np.sin(t*F[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*F[3])*np.exp(-d[3]*t)

plt.plot(x, y, 'k', linewidth=.1)
plt.show()

y = A[0]*np.sin(t*F[0])*np.exp(-d[0]*t) + A[1]*np.sin(t*F[1])*np.exp(-d[1]*t)
x = A[2]*np.sin(t*F[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*F[3])*np.exp(-d[3]*t)

plt.plot(x, y, 'k', linewidth=.1)
plt.axis('off')
plt.show()

y = A[1]*np.sin(t*F[1])*np.exp(-d[1]*t) + A[3]*np.sin(t*F[3])*np.exp(-d[3]*t)
x = A[2]*np.sin(t*F[2])*np.exp(-d[2]*t) + A[0]*np.sin(t*F[0])*np.exp(-d[0]*t)

plt.plot(x, y, linewidth=.1)
plt.show()

x = A[1]*np.sin(t*F[1])*np.exp(-d[1]*t) + A[3]*np.sin(t*F[3])*np.exp(-d[3]*t)
y = A[2]*np.sin(t*F[2])*np.exp(-d[2]*t) + A[0]*np.sin(t*F[0])*np.exp(-d[0]*t)

plt.plot(x, y)
plt.axis('off')
plt.show()

print(x,y)
```
